# Parallelogram Law  
*Machine Learning Made Easy and Visual with Python*

## Initialization

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Define vectors

```python
a_vec = np.array([4, 1])  # Vector a
b_vec = np.array([1, 2])  # Vector b
```

## Compute vector sum

```python
a_plus_b = a_vec + b_vec  # Result of vector addition a + b
```

## Visualization

```python
plt.figure(figsize=(6,6))  # Create a square plot for equal scaling

# Plot vector a
plt.quiver(0, 0, 
           a_vec[0], a_vec[1], 
           angles='xy', scale_units='xy', scale=1, 
           color='r', label="a")

# Plot vector b
plt.quiver(0, 0, 
           b_vec[0], b_vec[1], 
           angles='xy', scale_units='xy', scale=1, 
           color='b', label="b")

# Plot vector a + b (the diagonal of the parallelogram)
plt.quiver(0, 0, 
           a_plus_b[0], a_plus_b[1], 
           angles='xy', scale_units='xy', scale=1, 
           color='pink', label="a + b")

# Draw the remaining sides of the parallelogram
plt.plot([b_vec[0], a_plus_b[0]], 
         [b_vec[1], a_plus_b[1]], 'k--')  # From tip of b to tip of a + b

plt.plot([a_vec[0], a_plus_b[0]], 
         [a_vec[1], a_plus_b[1]], 'k--')  # From tip of a to tip of a + b

# Set plot appearance
plt.gca().set_aspect('equal', adjustable='box')  # Keep equal scaling for x and y
plt.xlim(0, 5)  # Limit x-axis range
plt.ylim(0, 5)  # Limit y-axis range
plt.axhline(0, color='black', linewidth=0.5)  # Draw x-axis
plt.axvline(0, color='black', linewidth=0.5)  # Draw y-axis
plt.grid(color='gray', alpha=0.8, linestyle='-', linewidth=0.25)  # Add grid lines

plt.legend()  # Display legend
```
