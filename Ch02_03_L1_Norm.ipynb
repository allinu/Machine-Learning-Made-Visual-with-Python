# L1 Norm Contour Lines  
*Machine Learning Made Easy and Visual with Python*

## Initialization

```python
import numpy as np
import matplotlib.pyplot as plt 
```

## Generate data grid

```python
xx, yy = np.meshgrid(np.linspace(-8, 8, 1000), 
                     np.linspace(-8, 8, 1000))  # Create a fine grid for contour visualization
```

## Compute the L1 norm

```python
zz = np.abs(xx) + np.abs(yy)  # L1 norm formula: |x1| + |x2|
```

## Define standard basis vectors

```python
e1 = np.array([1, 0])  # Unit vector along the x1-axis
e2 = np.array([0, 1])  # Unit vector along the x2-axis
```

## Define vector colors

```python
color_e1 = [0, 0, 0]  # Black for e1
color_e2 = [0, 0, 0]  # Black for e2
```

## Visualization

```python
plt.figure(figsize=(6, 6))  # Create a square plot for equal scaling

# Plot standard basis vectors e1 and e2
plt.quiver(0, 0, e1[0], e1[1], 
           angles='xy', scale_units='xy', 
           scale=1, color=color_e1, 
           label='e1', zorder=1000)

plt.quiver(0, 0, e2[0], e2[1], 
           angles='xy', scale_units='xy', 
           scale=1, color=color_e2, 
           label='e2', zorder=1000)

# Plot L1 norm contour lines
# Each contour line represents all points with the same L1 norm value.
# Geometrically, these contours form diamond shapes centered at the origin.
contour = plt.contour(xx, yy, zz,
                      levels=np.arange(1, 9), 
                      cmap='RdYlBu_r')  

# Add colorbar to indicate L1 norm magnitude
cbar = plt.colorbar(contour)
cbar.set_label("L1 Norm (|x1| + |x2|)")

# Set axis limits
plt.xlim(-8, 8)
plt.ylim(-8, 8)

# Draw coordinate axes and grid lines
plt.axhline(0, color='black', linewidth=0.5)  # Horizontal axis
plt.axvline(0, color='black', linewidth=0.5)  # Vertical axis
plt.xticks(np.arange(-8, 9, 2))
plt.yticks(np.arange(-8, 9, 2))
plt.grid(True, which='both', linestyle='--', linewidth=0.5, color='0.8')

# Keep equal scaling for both axes
plt.gca().set_aspect('equal', adjustable='box')
plt.title("L1 Norm Contours and Basis Vectors")
plt.legend()
```
