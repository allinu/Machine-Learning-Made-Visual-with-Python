# Simple Linear Reression
*Machine Learning Made Easy and Visual with Python*

## Initialization

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

## Data preparation

```python
x = np.array([1,2,3,5,6,7,8,9]).reshape(-1,1)   # Independent variable (feature) x
y = np.array([3,1,4,6,5,8,7,8]).reshape(-1,1)   # Dependent variable (target) y
```

## Build a simple linear regression model

```python
model = LinearRegression()         # Initialize the Linear Regression model
model.fit(x, y)                    # Fit the model to training data (estimate coefficients)
```

## Retrieve regression coefficients and intercept

```python
b1 = model.coef_                   # Slope (shape: (1, 1))
b0 = model.intercept_              # Intercept (shape: (1,))
```

```python
print("Parameter vector b =", np.vstack([b0, b1]))  # Compare with manually derived parameters
```

## Compute predicted values on the training set

```python
y_pred = model.predict(x)          # Predicted values ŷ for each training sample
```

## Predict new values over a continuous range of x

```python
x_array = np.linspace(0, 10).reshape(-1,1)       # Generate evenly spaced points between 0 and 10
y_array_pred = model.predict(x_array)             # Predict corresponding y values
```

## Compute the residual (error) vector

```python
error = y - y_pred                  # Residuals = observed y minus predicted ŷ
```

## Compute the squared L2 norm of the residual vector

```python
squared_L2_norm = error.T @ error   # Equivalent to sum of squared residuals (SSR)
print("Sum of squared residuals =", squared_L2_norm)
```

## Visualization

```python
fig, ax = plt.subplots(figsize=(5,5))

# Plot observed sample data
ax.scatter(x, y, marker='x', color='b', label='Observed values')

# Plot the fitted regression line
ax.plot(x_array, y_array_pred, color='r', label='Regression line')

# Plot predicted points on the regression line
ax.scatter(x, y_pred, marker='x', color='r', label='Predicted values')

# Plot residuals as vertical orange segments between observed and predicted points
ax.plot(
    ([i for i in x.squeeze()], [i for i in x.squeeze()]),
    ([j for j in y_pred.squeeze()], [j for j in y.squeeze()]),
    c='#FFC000', alpha=0.5
)

# Axis labels and appearance settings
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.set_xticks(np.arange(11))
ax.set_yticks(np.arange(11))
ax.grid(True, c='0.8')
ax.set_aspect('equal', 'box')
ax.legend()
```
