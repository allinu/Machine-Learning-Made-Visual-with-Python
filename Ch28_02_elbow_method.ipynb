# Elbow Method
*Machine Learning Made Easy and Visual with Python*

##  Initialization 

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import numpy as np
import seaborn as sns
from sklearn import datasets
from sklearn.cluster import KMeans
```

##  Load the Iris dataset 

```python
iris = datasets.load_iris()

# Use only the first two features: sepal length and sepal width
X_train = iris.data[:, :2]

# True labels (for reference, not used in clustering)
y_train = iris.target
```

##  K-Means Clustering 

```python
# Store the distortion (sum of squared distances to the nearest cluster center)
distortions = []
for i in range(1, 11):
    # Initialize K-Means model with i clusters
    km = KMeans(
        n_clusters=i,        # number of clusters
        init='random',       # random initialization of centroids
        n_init=10,           # number of times the algorithm will be run with different centroid seeds
        max_iter=300,        # maximum number of iterations per run
        tol=1e-04,           # tolerance for convergence
        random_state=0       # ensures reproducibility
    )
    # Fit the model to the data
    km.fit(X_train)
    # Append the distortion (inertia) to the list
    distortions.append(km.inertia_)
```

##  Visualization: Elbow Method 

```python
fig, ax = plt.subplots()

# Plot number of clusters vs. distortion
plt.plot(range(1, 11), distortions, marker='x')
ax.grid(linestyle='--', linewidth=0.25, color=[0.5, 0.5, 0.5])

# Label axes
plt.xlabel('Number of Clusters', fontsize=12)
plt.ylabel('Distortion (Inertia)', fontsize=12)

# Set x-axis ticks
ax.set_xticks(range(1, 11))

plt.title('Elbow Method for Optimal k in K-Means', fontsize=14)
plt.show()
```
