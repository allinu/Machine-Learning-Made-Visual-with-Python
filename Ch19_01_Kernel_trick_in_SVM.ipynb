## Initialization

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_moons, make_circles, make_classification
from sklearn import svm, datasets
from matplotlib import cm
```

## Custom function to generate mesh grid for plotting decision boundaries

```python
def make_meshgrid(x, y, h=.02):
    # Determine min and max values for the grid with padding
    x_min, x_max = x.min() - 1, x.max() + 1
    y_min, y_max = y.min() - 1, y.max() + 1
    # Create 2D grid coordinates
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    return xx, yy
```

## Generate synthetic datasets

```python
X, y = make_classification(n_features=2, 
                           n_redundant=0, 
                           n_informative=2,
                           random_state=1, 
                           n_clusters_per_class=1)

# Add random noise to make dataset less perfectly linear
rng = np.random.RandomState(2)
X += 2 * rng.uniform(size=X.shape)
linearly_separable = (X, y)

# Define multiple datasets for experimentation
datasets = [linearly_separable,
            make_moons(noise=0.3, random_state=0),
            make_circles(noise=0.2, factor=0.5, random_state=1)]
```

## Iterate over datasets

```python
fig_name = 1

for ds_cnt, ds in enumerate(datasets):
    # Preprocess dataset: extract features and labels
    X, y = ds
    # Standardize features to zero mean and unit variance
    X = StandardScaler().fit_transform(X)

    C = 3  # SVM regularization parameter

    # Define multiple SVM models with different kernels
    models = (svm.SVC(kernel='linear', C=C),
              svm.SVC(kernel='poly', degree=2, gamma='auto', C=C),
              svm.SVC(kernel='poly', degree=3, gamma='auto', C=C),
              svm.SVC(kernel='rbf', gamma=0.7, C=C),
              svm.SVC(kernel='sigmoid', gamma=0.5, C=C))

    # Train all models on the dataset
    models = (clf.fit(X, y) for clf in models)
    
    # Titles for each SVM kernel visualization
    titles = ('Linear Kernel',
              'Polynomial Kernel, degree=2',
              'Polynomial Kernel, degree=3',
              'RBF Kernel',
              'Sigmoid Kernel')

    # Extract feature columns for plotting
    X0, X1 = X[:, 0], X[:, 1]
    xx, yy = make_meshgrid(X0, X1)
    
    # Iterate over each trained model
    for clf, title in zip(models, titles):
        
        # Compute decision function for all points in the grid
        Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
        Z = Z.reshape(xx.shape)
        Z_0 = clf.decision_function(X)
        
        # Create a new figure
        fig = plt.figure(figsize=(8, 4))
        ax = fig.add_subplot(1, 2, 1)
        
        # Plot 2D scatter plot of data points
        ax.scatter(X0, X1, c=y, cmap=plt.cm.coolwarm, s=20, edgecolors='w')

        # Highlight support vectors
        x1_sp_vec = clf.support_vectors_[:, 0]
        x2_sp_vec = clf.support_vectors_[:, 1]
        plt.scatter(x1_sp_vec, x2_sp_vec, s=50,
                    facecolors='none', edgecolors='k')
        
        # Plot decision boundary contours
        plt.contour(xx, yy, -Z, levels=30, cmap=plt.cm.RdYlBu_r, linewidths=0.5)
        plt.contour(xx, yy, -Z, [0], alpha=1, colors='#1B5A9C', linewidths=1.25)
        plt.contour(xx, yy, -Z, [-1, 1], alpha=1, colors='k', linewidths=1.25, linestyles='--')

        ax.set_xlabel('x1')
        ax.set_ylabel('x2')
        ax.set_xticks([])
        ax.set_yticks([])        
        ax.set_title(title)
        ax.set_box_aspect(1)
        
        # 3D decision surface plot
        ax = fig.add_subplot(1, 2, 2, projection='3d')
        norm = plt.Normalize(-Z.max(), -Z.min())
        colors = cm.RdYlBu_r(norm(-Z))
        
        surf = ax.plot_surface(xx, yy, -Z, linewidths=0.25,
                               facecolors=colors, shade=False, rstride=10, cstride=10)
        surf.set_facecolor((0, 0, 0, 0))

        # Overlay decision boundaries on the 3D surface
        ax.contour(xx, yy, -Z, [0], colors='#1B5A9C', linewidths=1.25)
        ax.contour(xx, yy, -Z, [-1, 1], colors='k', linewidths=1.25, linestyles='--')

        ax.set_xlabel('x1')
        ax.set_ylabel('x2')
        
        # Set orthographic projection and viewpoint
        ax.set_proj_type('ortho')
        ax.view_init(azim=-120, elev=30)
        ax.grid(False)
        ax.set_xticks([])
        ax.set_yticks([])
        ax.set_zticks([])
        ax.set_xlim(xx.min(), xx.max())
        ax.set_ylim(yy.min(), yy.max())

        ax.set_title(str(fig_name))
```
